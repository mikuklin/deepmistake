```python
import pandas as pd
from pathlib import Path
import csv
```

```python
dfs = []
for p in Path('../comedi_dataset_all/').glob('*/*.data'):
    df = pd.read_json(p)
    df.merge(pd.read_json(p.parent/p.name.replace('.data','.gold')), on='id', how='outer')
    dfs.append(df)
df = pd.concat(dfs, ignore_index=True)
assert ~df.isnull().any().any()
df['id'] = df['id'].str.split('.')
df['lang'] = df['id'].str[1].str.split('_').str[-1]
df['part'] = df['id'].str[0]
print(len(df))
df = df.sort_values(by=['lang','part','lemma'])
df.sample(10)
```

```python
key = ('sentence','start','end')
df = pd.concat([df.rename(columns={f'{c}{i}':c for c in key}).drop(columns=[f'{c}{3-i}' for c in key]) 
                for i in (1,2)], ignore_index=True)
df = df.drop(columns=('grp'))
print(len(df))
```

```python
df['positions'] = df.apply(lambda r: (r.start, r.end), axis=1)
df = df.rename(columns={'sentence':'context'})
df
```

```python
df.groupby('part').lemma.value_counts().to_csv('part-lemma-counts.tsv', sep='\t')
!column -t part-lemma-counts.tsv
```

# sanity check the specified positions of the target words

```python
df['wordat'] = df.apply(lambda r: r.context[slice(*r.positions)], axis=1)
df['wordat_ctxed'] = df.apply(lambda r: r.context[max(0,r.positions[0]-5):r.positions[1]+5 ], axis=1)
```

```python
forms = df.groupby(['lang','lemma']).wordat.apply(lambda s: 
                                        ', '.join(f'/{k}/{v}' for k,v in s.value_counts().to_dict().items())).reset_index()
pd.set_option('colwidth',300)
pd.set_option('display.max_rows',300)
forms.to_csv('forms.tsv', sep='\t')
forms.sample(20)
```

```python
# !pip install simplemma
from simplemma import lemmatize
```

```python
lang2code = {'norwegian':'nb','german':'de','spanish':'es', 'chinese':None, 
             'swedish':'sv', 'english':'en', 'russian':'ru'}
df['lang_code'] = df.lang.replace(lang2code)
```

```python
df['lemma_at'] = df.apply(lambda r: lemmatize(r.wordat, lang=r.lang_code) if r.lang_code else r.wordat, axis=1)
```

```python
q = df.groupby(['lang','lemma','part']).apply(lambda g: ' '.join(set(g.lemma_at))).reset_index()
```

```python
m0 = (q[0].str.split().str.len() == 1) & (q[0].str.split().str[0]==q.lemma.str.split('_').str[0])
print('1 normal form which is equal to lemma:', m0.mean())
m1 = q[0].str.split().str.len() > 1
print('More than 1 normal form found:',m1.mean())
m2 = (q[0].str.split().str.len() == 1) & (q[0].str.split().str[0]!=q.lemma.str.split('_').str[0])
print('1 normal form which is not lemma', m2.mean())
m3 = q.apply(lambda row: row.lemma.split('_')[0] in (row[0].split()), axis=1)

print(f'lemma is among normal forms: {m3.mean()}')
```

### Check first, last letters of wordat, a letter before and a letter after

```python
total = df.groupby(['lang','part'])[['wordat','lemma','context','positions']].nunique()
total
```

```python
df = df.sort_values(by=['lang','part','lemma'])
view_cols = ['lang','part','wordat','lemma', 'context', 'positions']
```

### first letter

```python
c = df.wordat.str[0]
c.value_counts(dropna=False)
```

```python
df[(~c.str.isalpha())][view_cols]
```

### last letter

```python
c = df.wordat.str[-1]
c.value_counts()  # last letter
```

```python
q = df[~c.str.isalpha()][view_cols]
q
```

```python
q.groupby(['lang','part']).nunique()/total
```

### symbol before the first letter

```python
# symbol before the first letter
c = df.apply(lambda r: ('^'+r.context)[r.positions[0]], axis=1)
c.value_counts().to_csv('symbol_before',sep='/')
!column -t <symbol_before
```

```python
q = df[c.str.isalpha()][view_cols]
q
```

```python
q.groupby(['lang','part']).nunique()/total
```

### symbol after the last letter

```python
c = df.apply(lambda r: (r.context+"$")[r.positions[1]], axis=1)
vc = c.value_counts()
vc.to_csv('symbol_before',sep='/')
!column -t <symbol_before
```

```python
q = df[c.str.isalpha()][view_cols]
q
```

```python
q.groupby(['lang','part']).nunique()/total
```
